Cleaning my FRED data next. For each table, prior to merging, I have made made the date the index, made it so that it has a full daily range within the min and max dates for each table, and forward filled the missing values. When I join them, the min date will be the lowest common date they have in common. Same for the max date. 

```python
df_GDP = pd.read_csv('data/GDPC1.csv')
df_GDP['observation_date'] = pd.to_datetime(df_GDP['observation_date'])
df_GDP.set_index('observation_date', inplace=True)
full_range = pd.date_range(start=df_GDP.index.min(), end=df_GDP.index.max(), freq='D')  
df_GDP = df_GDP.reindex(full_range)
df_GDP.index.name = 'observation_date' 
df_GDP.fillna(method='ffill', inplace=True)  
df_GDP.head()
```

```python
df_icsa = pd.read_csv('data/ICSA.csv')
df_icsa['observation_date'] = pd.to_datetime(df_icsa['observation_date'])
df_icsa.set_index('observation_date', inplace=True)
full_range = pd.date_range(start=df_icsa.index.min(), end=df_icsa.index.max(), freq='D')  
df_icsa = df_icsa.reindex(full_range)
df_icsa.index.name = 'observation_date' 
df_icsa.fillna(method='ffill', inplace=True)  
df_icsa.head()
```

```python
df_inflation = pd.read_csv('data/FPCPITOTLZGUSA.csv')
df_inflation = df_inflation.rename(columns={'FPCPITOTLZGUSA':'Inflation'})
df_inflation['observation_date'] = pd.to_datetime(df_inflation['observation_date'])
df_inflation.set_index('observation_date', inplace=True)
full_range = pd.date_range(start=df_inflation.index.min(), end=df_inflation.index.max(), freq='D')  
df_inflation = df_inflation.reindex(full_range)
df_inflation.index.name = 'observation_date' 
df_inflation.fillna(method='ffill', inplace=True) 
df_inflation.head()
```

```python
join_1 = pd.merge(df_GDP, df_icsa, how='inner', on='observation_date')
df_fred = pd.merge(join_1, df_inflation, how='inner', on='observation_date')
df_fred.head(10)
```

```python
df_unrate = pd.read_csv('data/UNRATE.csv')
df_unrate['observation_date'] = pd.to_datetime(df_unrate['observation_date'])
df_unrate.set_index('observation_date', inplace=True)
full_range = pd.date_range(start=df_unrate.index.min(), end=df_unrate.index.max(), freq='D')  
df_unrate = df_unrate.reindex(full_range)
df_unrate.index.name = 'observation_date' 
df_unrate.fillna(method='ffill', inplace=True) 
df_unrate.head(10)
```

```python
df_FRED = pd.merge(df_fred, df_unrate, how='inner', on='observation_date')
df_FRED.head()
```
